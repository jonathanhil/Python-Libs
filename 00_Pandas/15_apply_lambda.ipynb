```

# lambda funcction with condition
# lambda <arguments> : <statement1> if <condition> else <statement2>
```

```
lbd_2 = lambda number_1, number_2 : f"Numéro 1: {number_1}. Numméro 2: {number_2}"
```

```
lbd_2(34, 35)
```

```
# check if number is even or odd
lbd_3 = lambda number : f"{number} est pair" if number%2==0 else f"{number} est impair"
```

```
lbd_3(8)
```
