# `.Apply()`

`.apply()`est une fonction de pandas. Il est possible d'appliquer une fonction à TOUTES les valeurs d'UNE COLONNE en même temps. C'est un gain de temps incroyable, comparé aux boucles python.

Commençons par créer une fonction, que l'on appliquera ensuite à la colonne "Medal".

```python
def rank(medal):
  if medal == "Gold":
    return 1
  if medal == "Silver":
    return 2
  if medal == "Bronze":
    return 3


jo_df["Medal"]
```

```python
jo_df["Medal"].apply(rank)
```

On peut maintenant écraser les anciennes valeurs pour les nouvelles.

```python
jo_df['Medal'] = jo_df["Medal"].apply(rank)
jo_df
```

# `.iterrows()`

On souhaite créer deux colonnes:
* Une colonne qui récupère le nom
* Une colonne qui récupère le prénom
La fonction iterrows permet d'iterrer sur notre dataframe, elle nous retourne une série avec index

```python
empty_last_name = []

for (i, val) in jo_df.iterrows():
  empty_last_name.append(val["Athlete"].split(',')[0])
```

```python
jo_df["family_name"] = empty_last_name
```

```python
jo_df.head(30)
```

```python
empty_first_name = []

for (i, row) in jo_df.iterrows():
  if len(row["Athlete"].split(",")) == 2:
    empty_first_name.append(row["Athlete"].split(",")[1])
  elif len(row["Athlete"].split(",")) == 3:
    empty_first_name.append(row["Athlete"].split(",")[1])
  else:
    empty_first_name.append(" X ")
```

```python
jo_df["first_name"] = empty_first_name
jo_df
```

```python
empty_first_name.count(" X ")
```

```python
jo_df[jo_df["first_name"] == " X "]["Athlete"].unique()
```

```python
jo_df[jo_df["first_name"] == " X "]
```

# `.apply()` + `lambda`

On peut créer une fonction lambda, dans le but de créer un indicateur de comparaison pour la colonne `GDP per Capita`, le z_score.

```python
jo_df
```

```python
jo_df.isnull().sum()
```

```python
jo_df.dropna(subset = ["GDP per Capita"], inplace = True)
```

```python
jo_df.isnull().sum()
```

```python
jo_df["GDP z_score"] = jo_df["GDP per Capita"].apply(lambda y: (y - jo_df["GDP per Capita"].mean()) / jo_df["GDP per Capita"].std())
jo_df
```

```python
jo_df[jo_df["GDP z_score"] == jo_df["GDP z_score"].max()]
```

# `.pivot_table()`

Pandas reprend le pivot_table de Excel. C'est le même vocabulaire qui est utilisé.  
C'est l'occasion de répondre à des questions:

🤔 Le top 10 des pays qui ont remportés le plus de médailles en catégorie femme pour les JO d'hiver ?

```python
jo_df.head(1)
```

```python
jo_df.pivot_table(index = "Country", values = "Medal", columns = ("Season", "Gender"), aggfunc= "sum")\
  .sort_values(by= ("Winter", "Women"), ascending = False)\
  .head(10)
```

```python
jo_df.pivot_table(index = "Country", values = "Medal", columns = ("Season", "Gender"), aggfunc= "sum")\
  .sort_values(by= ("Winter", "Women"), ascending = False)\
  .head(10)\
  .plot(kind = "bar", stacked = True, figsize = (12,5));
```

🤔 Quel athlète féminine à remporté le plus de médailles olympiques ?

```python
jo_df.pivot_table(index="Athlete", values="Medal", columns=("Gender"), aggfunc="sum").sort_values(by= ("Women"), ascending = False).head(1)
```
